```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('comm_data_cleaned.csv')
```

```python
# Prevents rows from being truncated 
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
df.head(50)
```

```python
# Converts all negative values in 'Rating' column to 0.0

df['Rating'] = df['Rating'].clip(lower = 0)
df.head(50)
```

```python
df.describe()
```

```python
# Removes unecessary columns

df = df.drop(['comma', 'part_time', 'part_time_still'], axis=1)
df.describe()
```

```python
# Shows all job titles

df['job_simp'].value_counts()
```

```python
#hourly wage to annual 

df['min_salary'] = df.apply(lambda x: x.min_salary*2 if x.hourly ==1 else x.min_salary, axis =1)
df['max_salary'] = df.apply(lambda x: x.max_salary*2 if x.hourly ==1 else x.max_salary, axis =1)
```

```python
# Add's 0's to all min and max salaries 

df['min_salary'] = df.apply(lambda x: x.min_salary*1000, axis =1)
df['max_salary'] = df.apply(lambda x: x.max_salary*1000, axis =1)
```

```python
df.head(25)
```

```python
# Displays minimum salary for all jobs in a histogram

df.min_salary.hist()
```

```python
# Display max salary for all jobs in a histogram 

df.max_salary.hist()
```

```python
# Display average for all jobs in a histogram 

df.avg_salary.hist()
```

```python
df.columns
df_cat = df[['Location', 'Headquarters', 'Size','Type of ownership', 'Industry', 'Sector', 'Revenue', 'company_txt', 'job_state','same_state',
            'excel_yn', 'powerpoint_yn', 'micro_office_yn', 'analytics_yn', 'photoshop_yn', 'adobe_yn', 'adwords_yn']]
```

```python
for i in df_cat.columns:
    cat_num = df_cat[i].value_counts()
    print("graph for %s: total = %d" % (i, len(cat_num)))
    chart = sns.barplot(x=cat_num.index, y=cat_num)
    chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
    plt.show()
```

```python
for i in df_cat[['Location','Headquarters','company_txt']].columns:
    cat_num = df_cat[i].value_counts()[:20]
    print("graph for %s: total = %d" % (i, len(cat_num)))
    chart = sns.barplot(x=cat_num.index, y=cat_num)
    chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
    plt.show()
```
